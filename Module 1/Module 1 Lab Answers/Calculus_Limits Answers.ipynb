# Calculus - Limits

```python
import numpy as np
import matplotlib.pyplot as plt
# get matplot to show charts inline
%matplotlib inline
```

```python
# define a function, say f(x) = 1/x
def f(x):
    return 1/x
```

```python
# calculate value of f(x) for some values
print('f(4) = ',f(4))
```

```python
# calculate value of f(x) for 0
print('f(0) = ', f(0))
```

```python
# plot f(x) from a small number to a large number
x = np.arange(0.1,100,1)
y = f(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()
```

```python
# plot f(x) from a small number to a large number with logarithmic scale
x = np.arange(0.1,100,1)
y = f(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.yscale('log')
plt.grid()
plt.show()
```
